```python
##data3["Day_of_Week"]= data3["Day_of_Week"].replace(2, "Monday") 
##data3["Day_of_Week"]= data3["Day_of_Week"].replace(3, "Tuesday") 
##data3["Day_of_Week"]= data3["Day_of_Week"].replace(4, "Wednesday") 
##data3["Day_of_Week"]= data3["Day_of_Week"].replace(5, "Thursday") 
##data3["Day_of_Week"]= data3["Day_of_Week"].replace(6, "Friday") 
##data3["Day_of_Week"]= data3["Day_of_Week"].replace(7, "Saturday") 
##data3["Day_of_Week"]= data3["Day_of_Week"].replace(7, "Saturday") 
data3["Speed_limit"]= data3["Speed_limit"].replace(10, "Ten") 
data3["Speed_limit"]= data3["Speed_limit"].replace(20, "Twenty")
data3["Speed_limit"]= data3["Speed_limit"].replace(30, "Thirty") 
data3["Speed_limit"]= data3["Speed_limit"].replace(40, "Fourty") 
data3["Speed_limit"]= data3["Speed_limit"].replace(50, "Fifty") 
data3["Speed_limit"]= data3["Speed_limit"].replace(60, "Sixty") 
data3["Speed_limit"]= data3["Speed_limit"].replace(70, "Seventy") 
data3.head()
```

```python
dummyday = pd.get_dummies(data3['Day_of_Week'])
dummyroad = pd.get_dummies(data3['Road_Type'])
dummyspeed = pd.get_dummies(data3['Speed_limit'])
dummyphc = pd.get_dummies(data3['Pedestrian_Human_Control'])
dummyppf = pd.get_dummies(data3['Pedestrian_Physical_Facilities'])
dummylight = pd.get_dummies(data3['Light_Conditions'])
dummyweather = pd.get_dummies(data3['Weather_Conditions'])
```

```python
data4 = pd.concat([data3, dummyday, dummyroad, dummyspeed, dummyphc, dummyppf, dummylight, dummyweather], axis=1)
data4.head()
```

```python
import statsmodels.api as sm

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
y = data4['Number_of_Casualties']
```

```python
X = data4[['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Ten','Twenty','Fourty','Fifty','Sixty','Seventy','One way street','Roundabout','Single carriageway','Slip road','Control by other authorised person','Control by school crossing patrol','Central refuge','Footbridge or subway','Pedestrian phase at traffic signal junction','Zebra crossing','non-junction pedestrian crossing','Darkeness: No street lighting','Darkness: Street lighting unknown','Darkness: Street lights present and lit','Darkness: Street lights present but unlit','Fine with high winds','Fog or mist','Other','Raining with high winds','Raining without high winds','Snowing with high winds','Snowing without high winds']]

```

```python
X1 = sm.add_constant(X)
```

```python
results = sm.OLS(y,X1).fit()
```

```python
results.summary()
```

```python
#Removing unnecessary variables
X2 = data4[['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Twenty','Fourty','Fifty','Sixty','Seventy','One way street','Roundabout','Single carriageway','Slip road','Pedestrian phase at traffic signal junction','Zebra crossing','non-junction pedestrian crossing','Darkeness: No street lighting','Darkness: Street lights present and lit','Fog or mist','Raining without high winds']]
X3 = sm.add_constant(X2)
findings = sm.OLS(y,X3).fit()
findings.summary()
```
